```python
import pandas as pd
from biom import load_table
```

```python
# as fetched from the classifier results through the
# caret function named varImp, see 7-classifier.
data = [['592170', 100.00],
['594227',  92.51],
['355269',  89.11],
['364903',  88.10],
['368219',  87.26],
['1113256', 87.16],
['582379',  85.32],
['195651',  83.10],
['308081',  82.11],
['571458',  81.80],
['363507',  80.18],
['540982',  79.78],
['4070491', 79.00],
['1654477', 78.97],
['354283',  78.35],
['191593',  76.76],
['845273',  76.16],
['592866',  76.08],
['360238',  73.48],
['328283',  70.52]]


features = pd.DataFrame(data=data, columns=['greengenes', 'ScaledImportance'])
features.set_index('greengenes', inplace=True)

bt = load_table('otu_table.15000.no-diarrhea.biom')
```

```python
def funk(adelic):
    tax = bt.metadata(id=adelic.name, axis='observation')['taxonomy']
    tax = map(lambda x: x.split('__')[1], tax)
    return ' '.join(tax)
features['taxonomy'] = features.apply(funk, axis=1, reduce=False)
```
